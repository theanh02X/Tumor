```python
from torch.utils.data import Dataset
from torchvision import transforms
import os
from PIL import Image
import torch

class SAMFineTuneDataset(Dataset):
    def __init__(self, image_dir, mask_dir, image_size=256):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.image_files = sorted(os.listdir(image_dir))
        self.mask_files = sorted(os.listdir(mask_dir))
        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        image_path = os.path.join(self.image_dir, self.image_files[idx])
        mask_path = os.path.join(self.mask_dir, self.mask_files[idx])

        image = Image.open(image_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")

        image = self.transform(image)
        mask = self.transform(mask)
        mask = (mask > 0.5).float()  # Chuẩn hóa mask về nhị phân 0/1

        return image, mask
```

```python
import os

def prepare_dirs(tumor_type):
    base = f"/content/fine_tune_data_{tumor_type}"
    os.makedirs(f"{base}/images", exist_ok=True)
    os.makedirs(f"{base}/masks", exist_ok=True)
    return base
```

```python
from PIL import Image
import numpy as np
import torch
import torchvision.transforms as T
from segment_anything import SamPredictor, sam_model_registry
import cv2

def extract_roi_and_mask(img_path, model_yolo, predictor_sam, save_base, filename_prefix=""):
    image_bgr = cv2.imread(img_path)
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    # Detect bbox with YOLOv8
    results = model_yolo.predict(img_path, save=False, conf=0.25)[0].boxes.xyxy.cpu().numpy()

    for idx, box in enumerate(results):
        x1, y1, x2, y2 = map(int, box)
        # Expand box slightly (optional)
        pad = 10
        x1, y1 = max(0, x1 - pad), max(0, y1 - pad)
        x2, y2 = min(image_rgb.shape[1], x2 + pad), min(image_rgb.shape[0], y2 + pad)

        # Crop ROI and save
        roi = image_rgb[y1:y2, x1:x2]
        roi_path = f"{save_base}/images/{filename_prefix}_box{idx}.jpg"
        cv2.imwrite(roi_path, cv2.cvtColor(roi, cv2.COLOR_RGB2BGR))

        # Predict mask with SAM
        predictor_sam.set_image(image_rgb)
        input_box = np.array([x1, y1, x2, y2])
        masks, scores, _ = predictor_sam.predict(box=input_box[None, :], multimask_output=False)

        # Save mask (crop same region)
        mask_crop = masks[0][y1:y2, x1:x2].astype(np.uint8) * 255
        mask_path = f"{save_base}/masks/{filename_prefix}_box{idx}.png"
        cv2.imwrite(mask_path, mask_crop)
```

```python
import matplotlib.pyplot as plt

def visualize_mask_pipeline(img_path, model_yolo, predictor_sam, idx_to_show=0):
    # Đọc ảnh RGB
    image_bgr = cv2.imread(img_path)
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
    h, w, _ = image_rgb.shape

    # Phát hiện bằng YOLO
    results = model_yolo.predict(img_path, conf=0.25, save=False)[0]
    boxes = results.boxes.xyxy.cpu().numpy().astype(int)

    if len(boxes) == 0:
        print("Không tìm thấy bounding box nào.")
        return

    # Lấy bbox đầu tiên
    x1, y1, x2, y2 = boxes[idx_to_show]
    pad = 10
    x1, y1 = max(0, x1 - pad), max(0, y1 - pad)
    x2, y2 = min(w, x2 + pad), min(h, y2 + pad)

    # Cắt vùng ảnh crop
    roi = image_rgb[y1:y2, x1:x2]

    # Dự đoán mask bằng SAM
    predictor_sam.set_image(image_rgb)
    input_box = np.array([x1, y1, x2, y2])
    masks, scores, _ = predictor_sam.predict(box=input_box[None, :], multimask_output=False)
    mask_crop = masks[0][y1:y2, x1:x2].astype(np.uint8) * 255  # Cắt đúng vùng crop

    # Hiển thị ảnh
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))

    # 1. Ảnh gốc với bbox
    axs[0].imshow(image_rgb)
    axs[0].set_title("Ảnh gốc với bounding box")
    axs[0].add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, edgecolor='red', facecolor='none', lw=2))
    axs[0].axis("off")

    # 2. Ảnh crop
    axs[1].imshow(roi)
    axs[1].set_title("Vùng ROI crop từ YOLO")
    axs[1].axis("off")

    # 3. Mask từ vùng crop
    axs[2].imshow(mask_crop, cmap="gray")
    axs[2].set_title("Mask từ SAM (vùng crop)")
    axs[2].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
import os
import shutil

# Danh sách các thư mục gốc chứa các thư mục con images và masks
base_dirs = [
    "/content/drive/MyDrive/fine_tune_val/fine_tune_data_Glioma",
    "/content/drive/MyDrive/fine_tune_val/fine_tune_data_Meningioma",
    "/content/drive/MyDrive/fine_tune_val/fine_tune_data_Pituitary"
]

# Thư mục đích để gộp các file
destination_base_dir = "/content/drive/MyDrive/fine_tune_val/final_val"
destination_images_dir = os.path.join(destination_base_dir, "images")
destination_masks_dir = os.path.join(destination_base_dir, "masks")

# Tạo thư mục đích nếu chưa tồn tại
os.makedirs(destination_images_dir, exist_ok=True)
os.makedirs(destination_masks_dir, exist_ok=True)

# Gộp các file images
print("Copying image files...")
for base_dir in base_dirs:
    source_images_dir = os.path.join(base_dir, "images")
    if os.path.exists(source_images_dir):
        for filename in os.listdir(source_images_dir):
            source_path = os.path.join(source_images_dir, filename)
            destination_path = os.path.join(destination_images_dir, filename)
            if os.path.isfile(source_path):
                try:
                    shutil.copy2(source_path, destination_path)
                    # print(f"Copied image: {filename}") # Bỏ comment nếu muốn xem từng file
                except shutil.SameFileError:
                    print(f"Skipped image (same file): {filename}")
                except Exception as e:
                    print(f"Error copying image {filename}: {e}")
    else:
        print(f"Source images directory does not exist: {source_images_dir}")

print("\nCopying mask files...")
# Gộp các file masks
for base_dir in base_dirs:
    source_masks_dir = os.path.join(base_dir, "masks")
    if os.path.exists(source_masks_dir):
        for filename in os.listdir(source_masks_dir):
            source_path = os.path.join(source_masks_dir, filename)
            destination_path = os.path.join(destination_masks_dir, filename)
            if os.path.isfile(source_path):
                 try:
                    shutil.copy2(source_path, destination_path)
                    # print(f"Copied mask: {filename}") # Bỏ comment nếu muốn xem từng file
                 except shutil.SameFileError:
                     print(f"Skipped mask (same file): {filename}")
                 except Exception as e:
                     print(f"Error copying mask {filename}: {e}")
    else:
        print(f"Source masks directory does not exist: {source_masks_dir}")

print(f"\nAll image files have been copied to: {destination_images_dir}")
print(f"All mask files have been copied to: {destination_masks_dir}")
```
